#### Scipy:
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset given below.
#### Task:
* fitting it to the periodic function
* plot the fit

    Data
    * Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
    * Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

#### Expected Output:


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

thermax = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
thermin = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
month = np.arange(0,12)

# Pseudo Data Points
days = np.linspace(0,12, num = 365) # Dummy timeframe for curve fitting

def model(times, avg, amp, time_offset):
    omega = 2 * np.pi / times.max()
    return (avg + amp * np.cos(omega * (times + time_offset)))

res_max, cov_max = optimize.curve_fit(model, month, thermax, p0=[38, 13, 1])
res_min, cov_min = optimize.curve_fit(model, month, thermin, p0=[26, 9, 1])

plt.figure(figsize = (15,8))
plt.plot(month, thermax, 'ro') # Scatter-plot of data points for maximum temperature
plt.plot(days, model(days, * res_max), 'r-') # Fitting the non-linear curve using the model for maximum temperature
plt.plot(month, thermin, 'bo') # Scatter-plot of data points for minimum temperature
plt.plot(days, model(days, * res_min), 'b-') # Fitting the non-linear curve using the model for minimum temperature
plt.xlabel('Month')
plt.ylabel('Temperature(\u00B0C)')
```

### [Using trigonometric functions to model climate](https://niwa.co.nz/education-and-training/schools/resources/climate/modelling)
An equation that can be used to model these data is of the form: 
$$y = A cos B(x - C) + D$$
> where $A,B,C,D$ are constants, $y$ is the temperature in $°C$ and $x$ is the month (1–12).
> * $A$ = amplitude = $(ymax - ymin) / 2$
> * $B$ = $2π/12$
> * $C$ = units translated to the right
> * $D$ = $ymin + amplitude$ = units translated up

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

thermax = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
thermin = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
month = np.arange(0,12)

# Pseudo Data Points
days = np.linspace(0,12, num = 365) # Dummy timeframe for curve fitting

def model(times, A, B, C, D):
    
    return (A * np.cos(B * (times -C)) +D)

res_max, cov_max = optimize.curve_fit(model, month, thermax, p0=[13, 0.5, 1, 38], absolute_sigma=True)
res_min, cov_min = optimize.curve_fit(model, month, thermin,  p0=[9, 0.5, 1, 26], absolute_sigma=True)

plt.figure(figsize = (15,8))
plt.plot(month, thermax, 'ro') # Scatter-plot of data points for maximum temperature
plt.plot(days, model(days, * res_max), 'r-') # Fitting the non-linear curve using the model for maximum temperature
plt.plot(month, thermin, 'bo') # Scatter-plot of data points for minimum temperature
plt.plot(days, model(days, * res_min), 'b-') # Fitting the non-linear curve using the model for minimum temperature
plt.xlabel('Month')
plt.ylabel('Temperature(\u00B0C)')
```

    #### Charts to plot:

    2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic
```

## 1.

```python
pd.Categorical(titanic['sex'])
```

```python
cat = ['female', 'male']
size = [len(titanic.iloc[np.where(titanic['sex'] == val)]) for val in cat]
colors = ['lightcoral','lightskyblue']

plt.figure(figsize = (10,10))
plt.pie(size, labels = cat, colors = colors, autopct = '%.2f%%', shadow  = True)
```

## 2.

```python
plt.figure(figsize = (15, 8))
plt.scatter(titanic['age'], titanic['fare'], c = titanic['sex'].astype('category').cat.codes, cmap = 'rainbow_r')
```

```python
fig, ax = plt.subplots()

colors = {'female':'lightcoral', 'male':'green'}

grouped = titanic.groupby('sex')
for key, group in grouped:
    #print(key, group[['age', 'fare']])
    group.plot(ax = ax, kind='scatter', x = 'age', y = 'fare', label=key, color=colors[key], figsize = (15, 8))
```
